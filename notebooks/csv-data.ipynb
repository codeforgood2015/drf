```python
#Analyzing CSV data for DRF
#8 Jan 2015 | Benji
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime

#Graph Settings
pd.set_option('display.mpl_style', 'default')
# Always display all the columns
pd.set_option('display.width', 5000) 
pd.set_option('display.max_columns', 60) 

#Getting data (as dataframe)
#TODO: Change this path
df = pd.read_csv('/Users/bxie/programming/codeforgood/drf/data/twitter/tweet_activity_metrics.csv',
                parse_dates=['time'], index_col='time')

#removing empty columns
df = df.dropna(axis=1, how="all")

#Adding some columns about time
df['month'] = df.index.month
df['weekday'] = df.index.weekday
df['hour'] = df.index.hour

# df.describe()
df.head(5)
```

```python
#Graph of engagement and retweets by weekday
weekday = df[['impressions']].groupby(df['weekday']).aggregate(sum)
weekday.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
weekday.plot(kind='bar')

#Graph of engagement 
hour = df[['impressions']].groupby(df['hour']).aggregate(sum)
hour.plot(kind='bar')

#TIME IS DEPENDENT ON WHEN TWEET WAS POSTED, NOT WHEN VIEWED
```

```python
#Get Top 5 # of engagements
engage = df.sort()['engagements']
engage.value_counts()[:5]
```

```python
engage.plot(kind='bar')
```

```python
#Looking at only tweets that were seen/engaged with
engaged = df[df['impressions']>0][df['engagements']>0]
engaged.describe()
```

```python
df['weekday']=df.index.weekday
events = df[['weekday', 'engagement rate']]
weekday_counts = events.groupby('weekday').aggregate(sum)
# print weekday_counts
weekday_counts.index = ['Mon', 'Tues', 'Wed', 'Thurs', 'Fri', 'Sat', 'Sun']
weekday_counts.plot(kind='bar')
```
