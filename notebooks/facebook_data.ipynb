```python
from facepy import GraphAPI
import pandas as pd
import numpy as np
import datetime
#This comes from your Facebook and must be updated frequently.  Also, if you don't get it a certain way, it won't give you all the necessary permissions
oauth_access_token='CAAAAGqAx6bYBAAYb6EnArZChr6kxDdN1cHUDlt0SuyWjLt9TMqvFviNtRc8QA4lrceGn3MJ4DMspXXZBhGMdEIpp4ZAuR2HvnWflR2nESWCZALoUHkyBZCCLmhBmN1lbn96LUNHlpHB3IpoRo3cdh70NfL3lw4bjRVtjLgxTZAC0DyQM8jmEgOvrZBMOnLe2pAXLNmnfEwpRhfhMwUj3Qky'
#This just gets access
graph = GraphAPI(oauth_access_token)

#This gets the words from a post. Sometimes, there aren't any, in which case "No words could be found for this post" is returned. I'm not sure why this happens
def getWords(post):
    try:
        words=post['message']
    except KeyError:
        try:
            words=post['story']
        except KeyError:
            words="No words could be found for this post"
    return words

#This returns the number of likes a post has.
def getLikes(post):
    try:
        likes=len(post['likes']['data'])
    except KeyError:
        likes=0
    return likes

#This returns the number of comments a post has.
def getComments(post):
    try:
        comments=len(post['comments']['data'])
    except KeyError:
        comments=0
    return comments

#This returns the number of shares a post has.
def getShares(post):
    try:
        shares=post['shares']['count']
    except:
        shares=0
    return shares

#This returns the type of post according to the GraphAPI.  There are lots of types, and I am not sure what they all mean, so there is still work to be done here
def getType(post):
    try:
        postType=post['status_type']
    except KeyError:
        postType=post['type']
    return postType

#This gets the time a post was posted.  I haven't looked into adjusting for timezone yet.
def getTime(post):
    try:
        time=post['created_time']
    except KeyError:
        time='No time could be found for this post'
    return time

#This returns whether a post has a link in it.
def hasLink(post):
    words=getWords(post)
    hasLink=False
    for n in range(0,len(words)):
        if words[n:n+4]=='http':
            hasLink=True
    return hasLink
#This provides a convenient way to work with post data.
class easyPost:
    def __init__(self, post):
        self.words=getWords(post)
        self.type=getType(post)
        self.time=getTime(post)
        
        self.likes=getLikes(post)
        self.comments=getComments(post)
        self.shares=getShares(post)

        if self.type=='added_photo':
            self.photo=1
        else:
            self.photo=0

        if self.type=='shared_story':
            self.sharedStory=1
        else:
            self.sharedStory=0
            
        if self.type=='added_video':
            self.video=1
        else:
            self.video=0

        if self.type=='photo':
            self.updatePhoto=1
        else:
            self.updatePhoto=0

        if self.type=='mobile_status_update' or self.type=='status_update':
            self.status=1
        else:
            self.status=0

        if hasLink(post):
            self.link=1
        else:
            self.link=0
        
        if hasLink(post):
            self.link=1
        else:
            self.link=0
            
        if self.type=='created_note':
            self.note=1
        else:
            self.note=0
            
#This turns posts into easyPosts
def geteasyPosts(posts):
    easyPosts=[]
    try:
        for post in posts['posts']['data']:
            easyPosts=easyPosts+[easyPost(post)]
    except KeyError:
        for post in posts['data']:
            easyPosts=easyPosts+[easyPost(post)]
    return easyPosts

#This uses the paging of posts to get the next posts
def getNextPosts(posts):
    try:
        nextQuery=posts['posts']['paging']['next'][32:]
    except KeyError:
        nextQuery=posts['paging']['next'][32:]
    nextPosts=graph.get(nextQuery)
    return nextPosts

#This gets all posts from a page and turns them into easyPosts. It will continue getting posts until the limit is reached
def getAlleasyPosts(ID, limit=1000):
    posts=graph.get(ID+'/posts')
    easyPosts=geteasyPosts(posts)
    try:
        postArray=[getNextPosts(posts)]
        easyPosts=easyPosts+geteasyPosts(postArray[0])
        canContinue=True
    except KeyError:
        canContinue=False
    while canContinue:
        try:
            postArray=postArray+[getNextPosts(postArray[-1])]
            easyPosts=easyPosts+geteasyPosts(postArray[-1])
        except KeyError:
            canContinue=False
        if len(easyPosts)>=limit:
            canContinue=False
            easyPosts=easyPosts[0:limit]
    return easyPosts

#This gets the ID of a page by the page username i.e. facebook.com/{username}
def getID(username):
    results=graph.search(username,'page')
    return results['data'][0]['id']
#This dictionary can be used to keep track of ID numbers we need
ID={'DRF': '182413241850729'} 
```

```python
#Turns a string into an integer; works when string begins with 0
def toInt(intstr):
    newint=intstr
    while newint[0]=='0':
        newint=newint[1:]
    return int(newint)
#Converts the time that facebook gives into a datetime object
def todatetime(facebookTime):
    year=toInt(facebookTime[0:4])
    month=toInt(facebookTime[5:7])
    day=toInt(facebookTime[8:10])
    hour=toInt(facebookTime[11:13])
    minute=toInt(facebookTime[14:16])
    second=toInt(facebookTime[17:19])
    return datetime.datetime(year,month,day,hour,minute,second)

#Given a Facebook ID, puts all the necessary information into a dataframe
def get_data(ID, limit=1000):
    df = pd.DataFrame(columns=('time', 'type', 'text', 'likes', 'comments', 
                               'shares', 'hasLink', 'isPhoto', 'isVideo'), dtype=None)
    
    #populating dataframe
    easyPosts=getAlleasyPosts(ID,limit)
    for item in easyPosts:
        index = len(df)

        df.set_value(index, 'time', todatetime(item.time)-datetime.timedelta(hours=5)) #The hour in the time here is 5 greater than it is in East Coast Time
        df.set_value(index, 'text', item.words)
        df.set_value(index, 'type', item.type)
        df.set_value(index, 'likes', item.likes)
        df.set_value(index, 'comments', item.comments)
        df.set_value(index, 'shares', item.shares)
        df.set_value(index, 'hasLink', item.link)
        df.set_value(index, 'isPhoto', item.photo)
        df.set_value(index, 'isVideo', item.video)

    df['likes'] = df['likes'].astype(int)
    df['comments'] = df['comments'].astype(int)
    df['shares'] = df['shares'].astype(int)
    df['month'] = [dt.month for dt in df['time']]
    df['weekday'] = [dt.weekday() for dt in df['time']]
    df['hour'] = [dt.hour for dt in df['time']]
    
    
    return df
    
```
